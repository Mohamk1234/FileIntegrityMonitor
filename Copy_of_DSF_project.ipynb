# finding seasonal delays


```python
def top3_delayed_months(airport_flight, airport_code):
    # Assuming 'Date (MM/DD/YYYY)' column is in datetime format
    airport_flight['Date (MM/DD/YYYY)'] = pd.to_datetime(airport_flight['Date (MM/DD/YYYY)'])

    # Extract year from the date and create a new column
    airport_flight['Year'] = airport_flight['Date (MM/DD/YYYY)'].dt.year

    # Group by year and month and sum the weather delays
    monthly_weather_delays = airport_flight.groupby(['Year', airport_flight['Date (MM/DD/YYYY)'].dt.month])['Delay Weather (Minutes)'].sum()

    # Find top 3 delayed months for each year
    top3_delays = (
        monthly_weather_delays.groupby('Year', group_keys=False)
                              .apply(lambda x: x.nlargest(3))
                              .reset_index()
    )

    # Merge results of all years
    merged_results = top3_delays.groupby(['Date (MM/DD/YYYY)']).size().reset_index(name='Count')

    # Print the results
    print(f"Top 3 Weather Delayed Months for {airport_code}:")
    print(top3_delays)

    print(f"\nMerged Monthly Occurrence Counts for {airport_code}:")
    print(merged_results)

airports = {
    'ATL': ATL_flight,
    'DEN': DEN_flight,
    'JFK': JFK_flight,
    'SEA': SEA_flight,
    'DFW': DFW_flight,
    'ITO': ITO_flight,
    'LAX': LAX_flight,
    'MIA': MIA_flight,
    'ORD': ORD_flight,
    'SFO': SFO_flight
}

for airport_code, data in airports.items():
    top3_delayed_months(data, airport_code)
```

# Merging
Perform a join on flight and metar based on date and hour(scheduled dept for time) of the day if need be.

```python
def merger(flight,weather):
  flight["new date"] = pd.to_datetime(flight["Date (MM/DD/YYYY)"], format="%m/%d/%Y").dt.strftime("%Y-%m-%d")
  flight['time_stamp']=flight['new date']+" "+flight['Scheduled departure time']
  flight=flight[flight['Departure delay (Minutes)']>15]

  flight['time_stamp'] = pd.to_datetime(flight['time_stamp'])

  weather['valid_time_start'] = pd.to_datetime(weather['valid_time_start'])
  weather['valid_time_end'] = pd.to_datetime(weather['valid_time_end'])
  flight['time_stamp'] = pd.to_datetime(flight['time_stamp'])
  flight.sort_values(by=['time_stamp'], inplace=True)

  result = pd.merge_asof(
      flight, weather,
      left_on='time_stamp',
      right_on=['valid_time_start'],
      direction='nearest'
  )
  return result
```

These weather conditions collectively result in a higher number of flight delays during the summer season at DEN.

Denver experiences significant snowfall during the winter months, impacting airport operations and leading to delays. Mist and freezing fog contribute to reduced visibility, while freezing drizzle and snow add to the challenges. The diverse winter weather conditions make DEN susceptible to disruptions during this season.
